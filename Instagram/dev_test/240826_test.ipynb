```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from webdriver_manager.chrome import ChromeDriverManager
import time
import pandas as pd
from datetime import datetime, timedelta

# Î°úÍ∑∏Ïù∏ Ï†ïÎ≥¥
USERNAME = 'your_username'
PASSWORD = 'your_password'

# Chrome ÏòµÏÖò ÏÑ§Ï†ï
chrome_options = Options()
chrome_options.add_argument("--headless")
chrome_options.add_argument("--no-sandbox")
chrome_options.add_argument("--disable-dev-shm-usage")

# ÏõπÎìúÎùºÏù¥Î≤Ñ ÏÑ§Ï†ï
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=chrome_options)

def login_instagram():
    driver.get("https://www.instagram.com/accounts/login/")
    time.sleep(5)  # ÌéòÏù¥ÏßÄ Î°úÎìú ÎåÄÍ∏∞

    # Î°úÍ∑∏Ïù∏
    username_field = driver.find_element(By.NAME, "username")
    password_field = driver.find_element(By.NAME, "password")
    login_button = driver.find_element(By.XPATH, '//button[@type="submit"]')

    username_field.send_keys(USERNAME)
    password_field.send_keys(PASSWORD)
    login_button.click()
    time.sleep(5)  # Î°úÍ∑∏Ïù∏ Ï≤òÎ¶¨ ÎåÄÍ∏∞

def get_following_accounts():
    driver.get(f"https://www.instagram.com/{USERNAME}/")
    time.sleep(5)  # ÌéòÏù¥ÏßÄ Î°úÎìú ÎåÄÍ∏∞

    try:
        # ÌåîÎ°úÏö∞ Î™©Î°ù ÎßÅÌÅ¨
        following_link = driver.find_element(By.CSS_SELECTOR, 'a[href*="/following/"]')
        following_link.click()
        time.sleep(5)  # ÌéòÏù¥ÏßÄ Î°úÎìú ÎåÄÍ∏∞

        # ÌåîÎ°úÏö∞ Î™©Î°ù Ïä§ÌÅ¨Î°§
        following_list = []
        last_height = driver.execute_script("return document.body.scrollHeight")
        while True:
            driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
            time.sleep(3)
            new_height = driver.execute_script("return document.body.scrollHeight")
            if new_height == last_height:
                break
            last_height = new_height

        # ÌåîÎ°úÏö∞Ìïú Í≥ÑÏ†ï Î™©Î°ù Ï∂îÏ∂ú
        elements = driver.find_elements(By.CSS_SELECTOR, 'a.FPmhX.notranslate.MBL3Z')
        for element in elements:
            following_list.append(element.text)

        return following_list

    except Exception as e:
        print(f"ÌåîÎ°úÏö∞ Î™©Î°ù Í∞ÄÏ†∏Ïò§Í∏∞ Ïã§Ìå®: {e}")
        return []

def get_post_details(post_element):
    try:
        # Í≤åÏãúÍ∏Ä URL
        url = post_element.find_element(By.XPATH, './/a').get_attribute('href')
    except:
        url = "N/A"

    try:
        # Í≤åÏãúÍ∏Ä Î≥∏Î¨∏
        post_text = post_element.find_element(By.XPATH, './/div[contains(@class, "C4VMK")]/span').text
    except:
        post_text = "N/A"

    try:
        # Ï¢ãÏïÑÏöî Ïàò
        likes = post_element.find_element(By.XPATH, './/div[contains(@class, "Nm9Fw")]/button/span').text
    except:
        likes = "N/A"

    try:
        # Ìï¥ÏãúÌÉúÍ∑∏ Ï∂îÏ∂ú
        hashtags = [tag.text for tag in post_element.find_elements(By.XPATH, './/a[contains(@href, "/explore/tags/")]')]
    except:
        hashtags = []

    try:
        # Í≤åÏãúÍ∏Ä ÏóÖÎ°úÎìú ÎÇ†Ïßú
        post_date_element = post_element.find_element(By.XPATH, './/time')
        post_date = post_date_element.get_attribute('datetime')
        post_date = datetime.strptime(post_date, "%Y-%m-%dT%H:%M:%S.%fZ").strftime("%Y-%m-%d %H:%M:%S")
    except:
        post_date = "N/A"

    return {
        'url': url,
        'post_text': post_text,
        'likes': likes,
        'hashtags': ', '.join(hashtags),
        'post_date': post_date,
        'crawl_time': datetime.now().strftime("%Y-%m-%d %H:%M:%S")
    }

def get_recent_posts_from_account(account_name):
    account_url = f"https://www.instagram.com/{account_name}/"
    driver.get(account_url)
    time.sleep(5)  # ÌéòÏù¥ÏßÄ Î°úÎìú ÎåÄÍ∏∞

    # Ïä§ÌÅ¨Î°§ÌïòÏó¨ Í≤åÏãúÍ∏Ä Î°úÎìú
    last_height = driver.execute_script("return document.body.scrollHeight")
    while True:
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
        time.sleep(5)  # ÌéòÏù¥ÏßÄ Î°úÎìú ÎåÄÍ∏∞
        new_height = driver.execute_script("return document.body.scrollHeight")
        if new_height == last_height:
            break
        last_height = new_height

    # Í≤åÏãúÍ∏Ä Ï∂îÏ∂ú
    post_elements = driver.find_elements(By.XPATH, '//article//div[contains(@class, "v1Nh3")]/a')
    post_details = []

    for post in post_elements:
        post_element = post.find_element(By.XPATH, '..')
        post_info = get_post_details(post_element)
        
        # Í≤åÏãúÍ∏Ä ÎÇ†ÏßúÍ∞Ä ÏµúÍ∑º 1Ï£ºÏùº ÎÇ¥Ïù∏ÏßÄ ÌôïÏù∏
        post_date = datetime.strptime(post_info['post_date'], "%Y-%m-%d %H:%M:%S") if post_info['post_date'] != "N/A" else datetime.min
        if post_date >= datetime.now() - timedelta(days=7):
            post_details.append(post_info)

    # Îç∞Ïù¥ÌÑ∞ÌîÑÎ†àÏûÑÏúºÎ°ú Î≥ÄÌôò
    df = pd.DataFrame(post_details)
    df.to_csv(f'{account_name}.csv', index=False)
    print(f"Data for account '{account_name}' saved to '{account_name}.csv'")

# Î°úÍ∑∏Ïù∏ ÏàòÌñâ
login_instagram()

# ÌåîÎ°úÏö∞Ìïú Í≥ÑÏ†ï Î™©Î°ù Í∞ÄÏ†∏Ïò§Í∏∞
accounts = get_following_accounts()
print(f"Following accounts: {accounts}")

for account in accounts:
    get_recent_posts_from_account(account)

driver.quit()
```
