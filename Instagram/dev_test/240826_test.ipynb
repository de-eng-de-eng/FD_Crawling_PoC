```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from webdriver_manager.chrome import ChromeDriverManager
import time
import pandas as pd
from datetime import datetime, timedelta
import re


# Chrome 옵션 설정
chrome_options = Options()
chrome_options.add_argument("--headless")  # 백그라운드에서 실행

# 웹드라이버 설정
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=chrome_options)

def get_post_details(post_element):
    try:
        # 게시글 URL
        url = post_element.find_element(By.XPATH, './/a').get_attribute('href')
    except:
        url = "N/A"

    try:
        # 게시글 본문
        post_text = post_element.find_element(By.XPATH, './/div[contains(@class, "C4VMK")]/span').text
    except:
        post_text = "N/A"

    try:
        # 좋아요 수
        likes = post_element.find_element(By.XPATH, './/section[contains(@class, "EDfFK")]/div/div/span').text
    except:
        likes = "N/A"

    try:
        # 해시태그 추출
        hashtags = [tag.text for tag in post_element.find_elements(By.XPATH, './/a[contains(@href, "/explore/tags/")]')]
    except:
        hashtags = []

    try:
        # 게시글 업로드 날짜
        post_date_element = post_element.find_element(By.XPATH, './/time')
        post_date = post_date_element.get_attribute('datetime')
        post_date = datetime.strptime(post_date, "%Y-%m-%dT%H:%M:%S.%fZ").strftime("%Y-%m-%d %H:%M:%S")
    except:
        post_date = "N/A"

    return {
        'url': url,
        'post_text': post_text,
        'likes': likes,
        'hashtags': ', '.join(hashtags),
        'post_date': post_date,
        'crawl_time': datetime.now().strftime("%Y-%m-%d %H:%M:%S")
    }

def get_recent_posts_from_account(account_name):
    account_url = f"https://www.instagram.com/{account_name}/"
    driver.get(account_url)
    time.sleep(5)  # 페이지 로드 대기

    # 스크롤하여 게시글 로드
    for _ in range(3):  # 페이지를 여러 번 스크롤하여 더 많은 게시글 로드
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
        time.sleep(3)  # 페이지 로드 대기

    # 게시글 추출
    posts = driver.find_elements(By.XPATH, '//article//div[contains(@class, "v1Nh3")]/a')
    post_details = []

    for post in posts:
        post_element = post.find_element(By.XPATH, '..')
        post_info = get_post_details(post_element)
        
        # 게시글 날짜가 최근 1주일 내인지 확인
        post_date = datetime.strptime(post_info['post_date'], "%Y-%m-%d %H:%M:%S") if post_info['post_date'] != "N/A" else datetime.min
        if post_date >= datetime.now() - timedelta(days=7):
            post_details.append(post_info)

    # 데이터프레임으로 변환
    df = pd.DataFrame(post_details)
    df.to_csv(f'{account_name}.csv', index=False)
    print(f"Data for account '{account_name}' saved to '{account_name}.csv'")

# 계정 목록 (예: 사용자 계정)
accounts = ['seoul__nadri', 'sh.cookkook', 'teddy_zip_', 'hye_foodie_']

for account in accounts:
    get_recent_posts_from_account(account)

driver.quit()
```
