```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
import time
import pandas as pd
from datetime import datetime

# 1. ì›¹ë“œë¼ì´ë²„ ì„¤ì • ë° íŽ˜ì´ì§€ ì´ë™
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))

# ì¸ìŠ¤íƒ€ê·¸ëž¨ ë¦´ìŠ¤ URL
url = "https://www.instagram.com/reel/C_IGZmRyUU9/?utm_source=ig_web_copy_link&igsh=MzRlODBiNWFlZA=="
driver.get(url)
time.sleep(10)  # íŽ˜ì´ì§€ ë¡œë“œ ëŒ€ê¸°

# 2. ë°ì´í„° ì¶”ì¶œ
# ê³„ì •ëª… ì¶”ì¶œ
try:
    account_name = driver.find_element(By.XPATH, '//a[contains(@class, "notranslate")]').text
except:
    account_name = "N/A"


# ì¢‹ì•„ìš” ìˆ˜ ì¶”ì¶œ
try:
    # ì¢‹ì•„ìš” ìˆ˜ê°€ í¬í•¨ëœ span ìš”ì†Œë¥¼ ì°¾ìŠµë‹ˆë‹¤.
    likes_element = driver.find_element(By.XPATH, '//div[contains(text(), "likes")]')
    likes = likes_element.text.split(' ')[0]  # 'likes' í…ìŠ¤íŠ¸ ì•žì˜ ìˆ«ìžë§Œ ì¶”ì¶œ
except:
    likes = "N/A"
    

# í•´ì‹œíƒœê·¸ ì¶”ì¶œ
try:
    hashtags = [tag.text for tag in driver.find_elements(By.XPATH, '//a[contains(@href, "/explore/tags/")]')]
    hashtags = hashtags[:10]  # ìƒìœ„ 10ê°œ í•´ì‹œíƒœê·¸ë§Œ ê°€ì ¸ì˜¤ê¸°
except:
    hashtags = []

# ë³¸ë¬¸ í…ìŠ¤íŠ¸ ì¶”ì¶œ
try:
    post_text = driver.find_element(By.XPATH, '//div[@class="_a9zr"]').text
except:
    post_text = ""

# ê²Œì‹œ ë‚ ì§œ ì¶”ì¶œ
try:
    post_date_element = driver.find_element(By.XPATH, '//time')
    post_date = post_date_element.get_attribute('datetime')
    post_date = datetime.strptime(post_date, "%Y-%m-%dT%H:%M:%S.%fZ").strftime("%Y-%m-%d %H:%M:%S")  # ISO í˜•ì‹ì„ datetimeìœ¼ë¡œ ë³€í™˜
except:
    post_date = "N/A"

# í˜„ìž¬ í¬ë¡¤ë§ ì‹œê°„
crawl_time = datetime.now().strftime("%Y-%m-%d %H:%M:%S")

# 3. ê²°ê³¼ ì¶œë ¥ ë° CSV ì €ìž¥
data = {
    'account_name': account_name,
    'likes': likes,
    'hashtags': ', '.join(hashtags),
    'post_text': post_text,
    'post_date': post_date,
    'crawl_time': crawl_time,
    'post_url': url
}

# CSV ì €ìž¥
df = pd.DataFrame([data])
df.to_csv('{}.csv'.format(account_name), index=False)

print(f"Data saved to instagram_post_data.csv: {data}")


# ë“œë¼ì´ë²„ ì¢…ë£Œ
driver.quit()
```
